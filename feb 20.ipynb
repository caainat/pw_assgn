Explain GET and POST methods.
GET Method:

Used to request data from a server.

Parameters are sent in the URL.

Not secure for sensitive data.

Example:
http://example.com/user?name=Alice

POST Method:

Used to submit data to the server.

Data is sent in the body of the HTTP request.

More secure for sending sensitive information (like passwords).

Often used for form submissions or creating records.
